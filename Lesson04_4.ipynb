```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
ccfd = pd.read_csv('creditcard.csv')

ccfd.head()
```

```python
ccfd_class = ccfd.loc[:, 'Class']
```

```python
ccfd_class.value_counts().plot(kind="bar") 
```

```python
ccfd_class.value_counts().plot(kind="bar", logy=True)
```

```python
V1_Class_0 = ccfd.loc[ccfd['Class'] == 0, 'V1']
V1_Class_0
```

```python
V1_Class_1 = ccfd.loc[ccfd['Class'] == 1, 'V1']
V1_Class_1
```

```python
hist_info1 = plt.hist(V1_Class_0, bins=20, density=True, alpha=0.5, color='grey', label="Class 0")
hist_info2 = plt.hist(V1_Class_1, bins=20, density=True, alpha=0.5, color='red', label="Class 1")
plt.legend(loc="lower left", frameon=True)
plt.xlabel("V1")
plt.show()
```
